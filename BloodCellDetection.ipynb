# Import library

```python
import numpy as np 
import cv2 
import matplotlib.pyplot as plt 
```

# Import file

```python
# convert .bmp to grayscale file .png
bmp = cv2.imread('pic/bloodcell.bmp', cv2.IMREAD_GRAYSCALE)
cv2.imwrite('pic/BloodGray.png',bmp)
```

```python
# read .png file
img = cv2.imread('pic/BloodGray.png')
# convert conlor to HSV formath
hsv_img = cv2.cvtColor(img, cv2.COLOR_BGR2HSV)
```

# Cell detection

```python
# Black color in HSV is (0, 0, 0) 
# lower range in HSV
lower_range = (0, 0, 0) # lower is black color
# upper rangein HSV
upper_range = (0, 0, 75) # upper is black color in lightness 30%

# detect center of cell by color
mask = cv2.inRange(hsv_img, lower_range, upper_range)

# Show cell pictur VS center of cell detect 
plt.figure(figsize=(15,15))

plt.subplot(121),plt.imshow(img,cmap = 'gray')
plt.title('Original Image'), plt.xticks([]), plt.yticks([])

plt.subplot(122),plt.imshow(mask,cmap = 'gray')
plt.title('Center of cell detect'), plt.xticks([]), plt.yticks([])
```

```python
# find contouts in picture
contours,h = cv2.findContours(mask,1,2)

cell_count = 0

# count cell by contours zone
for cnt in contours:
    if cv2.contourArea(cnt) < 205: # contours zone < 205 is Blood Cell
        cv2.drawContours(img,[cnt],-1,(0,0,255),3)
        cell_count +=1

# show cell deterct 
plt.figure(figsize=(12,12))
plt.imshow(img,cmap = 'gray')
plt.title(f'Cell detect count : {cell_count}'), plt.xticks([]), plt.yticks([])

# seve output
cv2.imwrite('pic/BloodCellDection.png',img)
```
